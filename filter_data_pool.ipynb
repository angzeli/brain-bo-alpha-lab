# Data Pool Filter

This notebook filters combined master data-pool CSV files into focused subsets for later analysis.

It does not rank trials, plot results, or run modelling analysis.

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path

from data_pool_filter import filter_data_pool, expand_params_columns, apply_filters
```

## Choose Input CSV

Use a combined master data-pool CSV created by `combine_csv.ipynb`.

```python
INPUT_CSV = "combined_brain_bo_all_YYYY-MM-DD.csv"

Path(INPUT_CSV)
```

## Define Filters

Filters can use columns already present in the CSV, or fields expanded from the `params` column such as `template_type` and `neutralisation`.

```python
FILTERS = {
    "neutralisation": "Market",
}

# Examples:
# FILTERS = {"universe": "TOP3000"}
# FILTERS = {"neutralisation": "Market"}
# FILTERS = {"template_type": "price_reversion"}
# FILTERS = {"brain_rating": ["Average", "Good", "Excellent", "Spectacular"]}
# FILTERS = {"universe": "TOP3000", "neutralisation": "Industry"}
```

## Run Filter And Save

When `output_csv=None`, the utility creates a readable timestamped filename such as `subset_neutralisation_market_2026-05-06_203512.csv`.

```python
df_filtered = filter_data_pool(
    input_csv=INPUT_CSV,
    filters=FILTERS,
    output_csv=None,
)

print(f"Rows: {len(df_filtered)}")
print(f"Saved filtered subset to: {df_filtered.attrs.get('output_csv')}")

df_filtered.head()
```
